# creating an empty List and appending all texts inside tags

```python
def appendList():
    headings=[]
    for heads in h2:
        if len(heads.text)>0:
            headings.append('h2:'+heads.text)
    for heads in h3:
        if len(heads.text)>0:
            headings.append('h3:'+heads.text)
    for heads in p:
        if len(heads.text)>0:
            headings.append('p:'+heads.text)
    for heads in a:
        if len(heads.text)>0:
            headings.append('a:'+heads.text)
    return headings
```

```python
headings=appendList()
```

```python
headings
```

# creating file in write mode and writing list content

```python
def fileOperation(listdt):
    fo=open("LalamoveHongkong.txt","w",encoding='utf-8')
    for item in listdt:
        fo.write(str(item)+'\n')
```

# Question 1

# Here first left joined all customers with order table and calculated total spent by each customer, Here I considered all order amount in the current year, then assigned rank to each customer based on spent value, Rank value used because multiple customers may have same spent value. Then selected top 100 rank

```python
select * from 
(
    select x.customer_id,x.customer_name,x.spent,rank()over(order by x.spent desc)spent_rank
    from
    (
        Select t.customer_id,t.customer_name,nvl(sum(s.order_amount),0)spent
        from customers t
        left join orders s on s.customer_id=t.customer_id and extract(Year from s.order_date)=extract(Year from sysdate)
        group by t.customer_id,t.customer_name
    )x
)
where spent_rank<=100
order by customer_name
```

1.User sign up/ Login

    Where New customer registering by providing their basic information such as name,address,contact number etc.

    1) New vs Existing customer rate
    2) Bounce rate
           How quickly customers leaves website

2.Placing an order

    where Customer placed an order to ship a single or multiple products from location A to B
    1) Average order amount
    2) Order cancellation rate
    3) Number of Orders per given time period
    4) Average number of orders per customer
    
   Next KPIs are related to the business
   
3.Provisioning and storage

    where picking up all the products and keeping in the ware house before distribution.
    1) Average time between item reception to distribution
    2) Order preparation duration
            Average time required to prepare the product ready to distribute
    4) Storage cost
            cost required to store the product in ware house
            
4.Distribution

    1) Delivery time
            time range from when customer placed an order until they receives
    2) Cost per kg
    3) Order Accuracy
            Amount of orders processed with out any issue/damage
    4) Shipping time
            ratio of orders that have been shipped on or before requested time devided by total number of orders
    5) Transportation cost
    6) On time delivery
            Percentage of orders delivered within customer requested date
    7) Lead time
            It refers to how long the over all time between order placement and delivery
    8) Capacity Utlization
            during road transportation, how effectively utlizing vehicle capacity
    9) Average number of orders processed
            Number of orders processed within a given period (daily/weekly/monthly)
